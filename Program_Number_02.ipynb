## <font color='blue'>Problem Statement : 
### Simulate the smart vacuum cleaner agent that cleans rooms size of n * n. The Agent can move Up, Down, Left, Right. calculate performance each round

---

```python
import numpy as np
from datetime import datetime

# Taking input number of rooms
n = int(input('Enter the value of n : '))
rooms = np.random.randint(0, 2, (n, n))

# printing intial rooms
print('\nIntially Rooms')
print(rooms,'\n')

cleaned_rooms = 0
total_rooms = n*n

# logic for code
start = datetime.now()
for i in range(len(rooms)):
    for j in range(len(rooms[i])):
        if rooms[i][j] == 1:
            rooms[i][j] = 0
            cleaned_rooms += 1
            print(f'Rooms [{i},{j}] is cleaned.')
end = datetime.now()

# printing rooms after cleaning
print('\nAll rooms are cleaned.')
print(rooms,'\n')

# Performance and time calculation
performance = (cleaned_rooms / total_rooms) * 100 
time_taken = end - start
print(f'Perfomance of the program = {round(performance,2)} %')
print(f'Time taken = {time_taken.total_seconds()}')
```
